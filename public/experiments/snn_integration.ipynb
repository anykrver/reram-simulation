SNN on crossbar with simple input.

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path().resolve().parent))
import numpy as np
from src.crossbar import IdealCrossbar
from src.snn import PoissonEncoder, LIFNeuron, SNNNetwork

np.random.seed(42)
n_in, n_out, timesteps = 16, 8, 50
G = np.maximum(np.random.randn(n_in, n_out).astype(np.float32) * 0.1, 0)
crossbar = IdealCrossbar(n_in, n_out)
crossbar.set_conductance(G)

snn = SNNNetwork(n_in, n_out, crossbar_run=crossbar.run, timesteps=timesteps)
x = np.random.rand(n_in).astype(np.float32)
spikes, total = snn.forward(x, seed=42)
print("Output spikes shape:", spikes.shape)
print("Total spikes per neuron:", total)
print("Done.")
```
